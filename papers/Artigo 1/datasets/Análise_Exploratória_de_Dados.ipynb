## **Análise Exploratória de Dados de Pacientes com Câncer Colorretal**

Este código realiza uma análise exploratória de dados (EDA) com o objetivo de compreender melhor as características de um conjunto de dados relacionados a pacientes com câncer colorretal. O dataset contém informações demográficas, clínicas e tumorais, sendo utilizado para criar gráficos que ilustram aspectos importantes dos dados.

## Importação das Bibliotecas e Carregamento do Dataset

Nesta etapa, importamos as bibliotecas necessárias para análise e carregamos o dataset.

```python
# Importar bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Para suavizar as distribuições com linha de densidade

# Configurações de estilo
sns.set_theme(style="whitegrid")  # Configurações do seaborn

# Carregar o arquivo CSV
df = pd.read_csv('https://raw.githubusercontent.com/aureavaleria/Reprodu-o/main/export.csv')

# Configurar o tamanho da figura
plt.figure(figsize=(10, 8))
```

##Visualização da Distribuição de Variáveis Clínicas e Demográficas de Pacientes com Câncer Colorretal

```python
# Configurar o tamanho da figura maior para os gráficos
plt.figure(figsize=(16, 12))  # Aumentando a largura e altura da figura

# Gráfico 1 - Distribuição Etária com Porcentagem
plt.subplot(2, 2, 1)
age_counts = df['Age recode with <1 year olds'].value_counts(normalize=True).sort_index() * 100
age_counts.plot(kind='bar', color='dodgerblue')
plt.title('Distribuição Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Porcentagem (%)')
for i, v in enumerate(age_counts):
    plt.text(i, v + 0.5, f"{v:.2f}%", ha='center', color='black')

# Gráfico 2 - Distribuição por Sexo com Porcentagem
plt.subplot(2, 2, 2)
sex_counts = df['Sex'].value_counts(normalize=True) * 100
sex_counts.plot(kind='bar', color='skyblue')
plt.title('Distribuição por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Porcentagem (%)')
for i, v in enumerate(sex_counts):
    plt.text(i, v + 0.5, f"{v:.2f}%", ha='center', color='black')

# Gráfico 3 - Distribuição do Local Primário (Top 10) com Porcentagem
plt.subplot(2, 2, 3)
primary_site_counts = df['Primary Site'].value_counts(normalize=True).nlargest(10) * 100
primary_site_counts.plot(kind='bar', color='lightgreen')
plt.title('Distribuição do Local Primário (Top 10)')
plt.xlabel('Código do Local Primário')
plt.ylabel('Porcentagem (%)')
for i, v in enumerate(primary_site_counts):
    plt.text(i, v + 0.5, f"{v:.2f}%", ha='center', color='black')

# Gráfico 4 - Distribuição do Tamanho do Tumor com Porcentagem
plt.subplot(2, 2, 4)
plt.hist(df['CS tumor size (2004-2015)'], bins=30, color='salmon', weights=[100 / len(df)] * len(df))
plt.title('Distribuição do Tamanho do Tumor')
plt.xlabel('Tamanho do Tumor')
plt.ylabel('Porcentagem (%)')

# Adicionar linha de densidade para suavizar a visualização da distribuição
sns.kdeplot(df['CS tumor size (2004-2015)'], color='red', ax=plt.gca(), fill=True, alpha=0.2)

# Ajustar layout para melhor visualização
plt.tight_layout()

# Exibir todos os gráficos
plt.show()
```

## Visualização da Distribuição de Metástases no Pulmão e no Fígado

```python
# Configurar o tamanho da figura para os gráficos de metástase
plt.figure(figsize=(15, 5))

# Gráfico 5: Distribuição de Metástase no Pulmão com Porcentagem
plt.subplot(1, 3, 1)
lung_counts = df['SEER Combined Mets at DX-lung (2010+)'].value_counts(normalize=True) * 100
lung_counts.plot(kind='bar', color='skyblue')
plt.title('Distribuição de Metástase no Pulmão')
plt.xlabel('Metástase no Pulmão')
plt.ylabel('Porcentagem (%)')
for i, v in enumerate(lung_counts):
    plt.text(i, v + 1, f"{v:.2f}%", ha='center', color='black')

# Gráfico 6: Distribuição de Metástase no Fígado com Porcentagem
plt.subplot(1, 3, 2)
liver_counts = df['SEER Combined Mets at DX-liver (2010+)'].value_counts(normalize=True) * 100
liver_counts.plot(kind='bar', color='salmon')
plt.title('Distribuição de Metástase no Fígado')
plt.xlabel('Metástase no Fígado')
plt.ylabel('Porcentagem (%)')
for i, v in enumerate(liver_counts):
    plt.text(i, v + 1, f"{v:.2f}%", ha='center', color='black')
```

## Distribuição das Classificações TN do Sistema AJCC (7ª Edição)

```python
# Gráficos de distribuição para as classificações TN

plt.figure(figsize=(15, 5))

# Gráfico T: Tamanho do Tumor
plt.subplot(1, 3, 1)
t_counts = df['Derived AJCC T, 7th ed (2010-2015)'].value_counts(normalize=True).sort_index() * 100
t_counts.plot(kind='bar', color='yellow')
plt.title('Distribuição de T (Tamanho do Tumor)')
plt.xlabel('Classificação T')
plt.ylabel('Porcentagem (%)')
for i, v in enumerate(t_counts):
    plt.text(i, v + 0.5, f"{v:.2f}%", ha='center', color='black')

# Gráfico N: Linfáticos
plt.subplot(1, 3, 2)
n_counts = df['Derived AJCC N, 7th ed (2010-2015)'].value_counts(normalize=True).sort_index() * 100
n_counts.plot(kind='bar', color='salmon')
plt.title('Distribuição de N (Linfáticos)')
plt.xlabel('Classificação N')
plt.ylabel('Porcentagem (%)')
for i, v in enumerate(n_counts):
    plt.text(i, v + 0.5, f"{v:.2f}%", ha='center', color='black')


# Ajustar layout para melhor visualização
plt.tight_layout()
plt.show()
```
